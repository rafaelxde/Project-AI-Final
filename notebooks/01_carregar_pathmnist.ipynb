```python
from medmnist import PathMNIST
import matplotlib.pyplot as plt
import numpy as np

print("Bibliotecas importadas com sucesso!")
```

```python
train_set = PathMNIST(split="train", size=224, download=True)
val_set = PathMNIST(split="val", size=224, download=True)
test_set = PathMNIST(split="test", size=224, download=True)

print("Treino:", len(train_set))
print("Validação:", len(val_set))
print("Teste:", len(test_set))
```

```python
img, label = train_set[0]

print("Label:", label)
print("Tipo da imagem:", type(img))

plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
fig, axes = plt.subplots(2, 5, figsize=(12, 5))

for i, ax in enumerate(axes.flat):
    img, label = train_set[i]
    ax.imshow(img)
    ax.set_title(f"Classe: {label[0]}")
    ax.axis("off")

plt.show()
```

```python
print("Quantidade de imagens de treino:", len(train_set))
print("Quantidade de imagens de validação:", len(val_set))
print("Quantidade de imagens de teste:", len(test_set))

img, label = train_set[0]

print("Tamanho da imagem:", img.size)
print("Classe da primeira imagem:", label)
```
