```python
from medmnist import PathMNIST
import matplotlib.pyplot as plt
import numpy as np

print("Bibliotecas importadas com sucesso!")
```

```python
from medmnist import PathMNIST

# Visualização exploratória em 28x28.
# As etapas avançadas usam o sample real 224x224.
train_set = PathMNIST(split="train", size=28, download=True)
val_set = PathMNIST(split="val", size=28, download=True)

print("Treino:", len(train_set))
print("Validação:", len(val_set))
```

```python
img, label = train_set[0]

print("Label:", label)
print("Tipo da imagem:", type(img))

plt.imshow(img)
plt.axis("off")
plt.show()
```

## Visualização exploratória em 28×28

Este notebook utiliza a versão 28×28 apenas para visualização inicial e para reduzir o uso de memória no ambiente local. As etapas avançadas do projeto utilizam imagens reais da versão oficial 224×224, sem redimensionamento de imagens 28×28.

```python
fig, axes = plt.subplots(2, 5, figsize=(12, 5))

for i, ax in enumerate(axes.flat):
    img, label = train_set[i]
    ax.imshow(img)
    ax.set_title(f"Classe: {label[0]}")
    ax.axis("off")

plt.show()
```
